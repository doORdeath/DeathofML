## Matplotlib 의 Patches에 대해 알아보자!
### patches는 원,삼각형, 등 기하학적 형상을 그리는데 사용한다.

### 1. 원 그리기

```python
%matplotlib inline
import matplotlib.pyplot as plt

def create_circle():
    # (x,y) - 중심 좌표
    # 타원(Ellipse), 다각원(Polygon), 직사각형(Rectangle)패치도 가능.
    circle = plt.Circle((0,0),radius=0.05)
    return circle

def show_shape(patch):
    #현재 Axes에 대한 참조를 얻는다.
    ax=plt.gca()
    ax.add_patch(patch)
    plt.axis('scaled')
    plt.show()

if __name__ == '__main__':
    c= create_circle()
    show_shape(c)
```

### 2. 움직이는 그림 만들기 , Jupyter에서 움직이는게 안나옴 ㅠ

```python
from matplotlib import pyplot as plt
from matplotlib import animation

def update_radius(i,circle):
    circle.radius = i *0.5
    return circle
def create_animation():
    #Figure객체의 참조값을 받아온다.
    fig = plt.gcf()
    #-10 ~ 10 의 범위를 가치는 축생성
    ax = plt.axes(xlim=(-10,10),ylim=(-10,10))
    ax.set_aspect('equal')
    #원 추가
    circle = create_circle()
    ax.add_patch(circle)
    
    anim = animation.FuncAnimation(fig, update_radius,fargs=(circle,),frames=30,interval=5,repeat=True)
    plt.title('Simple Cirle Animation')
    plt.show()
    
if __name__ == '__main__':
    create_animation()
```

### 3. 시어핀스키의 삼각형 그리기

```python
from matplotlib import pyplot as plt
import random

def make_points(curXY):
    curChoice = random.randrange(1,4)
    if curChoice is 1:
        curXY[0] = curXY[0]*0.5
        curXY[1] = curXY[1]*0.5
    elif curChoice is 2:
        curXY[0] = curXY[0]*0.5+0.5
        curXY[1] = curXY[1]*0.5+0.5
    elif curChoice is 3:
        curXY[0] = curXY[0]*0.5+1
        curXY[1] = curXY[1]*0.5
    return curXY

#삼각형 점들의 집합
X=[]
Y=[]
cur=[0,0]
for x in range(10001):
    cur=make_points(cur)
    X.append(cur[0])
    Y.append(cur[1])

plt.plot(X,Y,'o')
plt.title('Sierpinski')
plt.xlabel('good')
plt.show()

```
