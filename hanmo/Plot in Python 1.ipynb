## pylab => matplotlib
### pylab의 plot은 대화식 셸에서 유용하지만, matplotlib의 pyplot을 이용하는 것이 더 효육적이다.

```python
#가장 처음 그래프
import matplotlib.pyplot

def create_graph():
    x_numbers=[1,2,3]
    y_numbers=[4,5,6]
    
    matplotlib.pyplot.plot(x_numbers,y_numbers)
    matplotlib.pyplot.show()
    
if __name__=='__main__':
    create_graph()
```

### 예) 포물선 운동을 그려보자!

#### 아쉽게도, range()는 부동소수점 간격으로 만들어 주지 않는다 ㅠ 

```python
# start - 시작
# fianl - ~까지
# increment - ~간격으로
def frange(start, final, increment):
    numbers=[]
    while start<final:
        numbers.append(start)
        start=start+increment
    return numbers
```

```python
from matplotlib import pyplot as plt
import math

#그래프를 그려주는 함수
def draw_graph(x,y):
    plt.plot(x,y)
    plt.xlabel('X-coordinate')
    plt.ylabel('Y-coordinate')
    plt.title('Projectle motion of a ball')
    
#포물선 운동 계산
# u = velocity(m/s)
def draw_tajectory(u,theta):
    theta = math.radians(theta)
    g=9.8
    
    #체공시간
    t_flight = 2*u*math.sin(theta)/g
    #시간 간격
    intervals = frange(0,t_flight,0.001)
    
    x=[]
    y=[]
    
    for t in intervals:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t - 0.5*g*t*t)
    draw_graph(x,y)

draw_tajectory(10,10)
draw_tajectory(20,20)
draw_tajectory(25,20)
plt.legend(['10m/s 10degree','20m/s 20degree','25m/s 20degree'])
```
